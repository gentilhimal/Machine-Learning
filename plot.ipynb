```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from numpy import genfromtxt

#reading required data from csv file
black = genfromtxt('housing.csv' , usecols=([12]))
price = genfromtxt('housing.csv', usecols=([13]))
room= genfromtxt('housing.csv', usecols=([5]))

plt.scatter(price,black)
plt.title('Housing Price')
plt.ylabel('Blacks in town')
plt.xlabel('Price')
plt.show()
```

```python

plt.scatter(price,room)
plt.ylabel('Rooms')
plt.xlabel('Price')
plt.show()
```

```python
multi = plt.gca(projection='3d')
multi.scatter(black,room,price)
multi.set_xlabel('black')
multi.set_ylabel('room')
multi.set_zlabel('price')
plt.show()
```
